```python
from habanero import Crossref
cr = Crossref()

# 通过DOI获取文献元数据
# doi = dois[0]

def get_metadata(doi):
  result = cr.works(ids=doi, format = "citeproc-json")

  # author = result.get("message", {}).get("author", None)
  title = result.get("message", {}).get("title", None)
  container = result.get("message", {}).get("container-title", None)
  # short_container = result.get("message", {}).get("short-container-title", None)
  # year = result.get("message", {}).get("published", {}).get("date-parts", None)[0][0]
  # volume = result.get("message", {}).get("volume", None)
  # page = result.get("message", {}).get("page", None)

  # return {"author": author, "title": title, "container": container, "short_container": short_container, "year": year, "volume": volume, "page": page, "doi": doi}
  return {"title": title[0], "publication": container[0]}
```

```python
result = {}
for ind, doi in enumerate(dois):
  # print(ind)
  try:
    result[doi.replace("/", "_")] = get_metadata(doi)
  except:
    print("error:", doi)
    continue
```

```python
import json
with open("articles.json", "w") as f:
  json.dump(result, f, ensure_ascii=False)
```

```python
import requests
from bs4 import BeautifulSoup
import os

save_folder = "./pdf"

if not os.path.exists(save_folder):
  os.makedirs(save_folder)

existing_files = os.listdir(save_folder)
# print(existing_files)

for ind, doi in enumerate(dois):
  filename = f'{doi.replace("/", "_")}.pdf'
  if filename in existing_files:
    continue
  print(ind)
  filename = os.path.join(save_folder, filename)
  try:
    content = requests.get(f"https://sci-hub.se/{doi}").text
    soup = BeautifulSoup(content)
    link = soup.select_one("embed").attrs["src"]
    if link.startswith("//"):
      link=f"https:{link}"
    elif link.startswith("/"):
      link=f"https://sci-hub.se{link}"

    response = requests.get(link, stream=True)
    with open(filename, "w") as f:
      pass
    with open(filename, "wb") as f:
      for chunk in response.iter_content(chunk_size=8192):
        f.write(chunk)
  except:
    print("error:", doi)
    continue
  # break
```

```python
import zipfile
import os

def zip_folder(folder_path, zip_filename):
    with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                file_path = os.path.join(root, file)
                arcname = os.path.relpath(file_path, folder_path)
                zipf.write(file_path, arcname)

# 使用示例
zip_folder(save_folder, 'articles.zip')
```
